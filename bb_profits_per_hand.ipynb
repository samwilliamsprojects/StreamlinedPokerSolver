```python
import pandas as pd
import re
import os 
```

```python
HANDS_DIRECTORY = 'PATH/TO/YOUR/HAND/HISTORY/DIRECTORY'
```

```python
def assign_seat():
    if 'ME' in line:
        my_seat = line.partition(":")[0]
        my_seat = my_seat[:4] + '+' + my_seat[5:]
        return my_seat
    else:
        return []
def blinds():
    total_blinds = 0
    if 'ME' in line:
        if 'Big Blind' in line:
            total_blinds = line.partition('$')[2] 
        elif 'Small Blind' in line:
            total_blinds = line.partition('$')[2]
    total_blinds = float(total_blinds)
    return total_blinds

def find_BB():
    if 'Big Blind' in line:
        value_line = line.partition('$')[2]
        bb_value = float(value_line)
        return bb_value
    
def card_assignment():
    hole_cards_raw = line[-8:-3]
    hole_cards_raw = hole_cards_raw[:2] + hole_cards_raw[3:]
    if hole_cards_raw[1] == hole_cards_raw[3]:
        hole_cards = hole_cards_raw[0] + hole_cards_raw[2] + 's'
    else:
        hole_cards = hole_cards_raw[0] + hole_cards_raw[2] + 'o'
    return hole_cards

def preflop_action():
    preflop_bet = 0
    if 'Folds' in line:
        return preflop_bet
    elif 'Checks' in line:
        return preflop_bet
    elif 'Bets' in line:
        value_line = line.partition('$')[2]
        total_split = re.findall('\d+', value_line)
        if len(total_split) > 1:
            preflop_bet = float(total_split[0] + '.' + total_split[1])
        else:
            preflop_bet = float(total_split[0])
        return preflop_bet
    elif 'Raises' in line:
        value_line = line.partition('$')[2]
        total_split = re.findall('\d+', value_line)
        preflop_bet = float(total_split[0] + '.' + total_split[1])
        return preflop_bet
    elif 'All-in(raise)' in line:
        value_line = line.partition('$')[2]
        total_split = re.findall('\d+', value_line)
        preflop_bet = float(total_split[0] + '.' + total_split[1])
        return preflop_bet
    elif 'All-in' in line:
        value_line = line.partition('$')[2]
        total_split = re.findall('\d+', value_line)
        if len(total_split) > 1:
            preflop_bet = float(total_split[0] + '.' + total_split[1])
        else:
            preflop_bet = float(total_split[0])
        return preflop_bet
    elif 'Calls' in line:
        value_line = line.partition('$')[2]
        total_split = re.findall('\d+', value_line)
        if len(total_split) > 1:
            preflop_bet = float(total_split[0] + '.' + total_split[1])
        else:
            preflop_bet = float(total_split[0])
        return preflop_bet
    else:
        return preflop_bet

def flop_action():
    flop_bet = 0
    if 'Folds' in line:
        return flop_bet
    elif 'Checks' in line:
        return flop_bet
    elif 'Bets' in line:
        value_line = line.partition('$')[2]
        total_split = re.findall('\d+', value_line)
        if len(total_split) > 1:
            flop_bet = float(total_split[0] + '.' + total_split[1])
        else:
            flop_bet = float(total_split[0])
        return flop_bet
    elif 'Raises' in line:
        value_line = line.partition('$')[2]
        total_split = re.findall('\d+', value_line)
        flop_bet = float(total_split[0] + '.' + total_split[1])
        return flop_bet
    elif 'All-in(raise)' in line:
        value_line = line.partition('$')[2]
        total_split = re.findall('\d+', value_line)
        flop_bet = float(total_split[0] + '.' + total_split[1])
        return flop_bet
    elif 'All-in' in line:
        value_line = line.partition('$')[2]
        total_split = re.findall('\d+', value_line)
        if len(total_split) > 1:
            flop_bet = float(total_split[0] + '.' + total_split[1])
        else:
            flop_bet = float(total_split[0])
        return flop_bet
    elif 'Calls' in line:
        value_line = line.partition('$')[2]
        total_split = re.findall('\d+', value_line)
        flop_bet = float(total_split[0] + '.' + total_split[1])
        return flop_bet

def turn_action():
    turn_bet = 0
    if 'Folds' in line:
        return turn_bet
    elif 'Checks' in line:
        return turn_bet
    elif 'Bets' in line:
        value_line = line.partition('$')[2]
        total_split = re.findall('\d+', value_line)
        if len(total_split) > 1:
            turn_bet = float(total_split[0] + '.' + total_split[1])
        else:
            turn_bet = float(total_split[0])
        return turn_bet
    elif 'Raises' in line:
        value_line = line.partition('$')[2]
        total_split = re.findall('\d+', value_line)
        turn_bet = float(total_split[0] + '.' + total_split[1])
        return turn_bet
    elif 'All-in(raise)' in line:
        value_line = line.partition('$')[2]
        total_split = re.findall('\d+', value_line)
        turn_bet = float(total_split[0] + '.' + total_split[1])
        return turn_bet
    elif 'All-in' in line:
        value_line = line.partition('$')[2]
        total_split = re.findall('\d+', value_line)
        if len(total_split) > 1:
            turn_bet = float(total_split[0] + '.' + total_split[1])
        else:
            turn_bet = float(total_split[0])
        return turn_bet
    elif 'Calls' in line:
        value_line = line.partition('$')[2]
        total_split = re.findall('\d+', value_line)
        if len(total_split) > 1:
            turn_bet = float(total_split[0] + '.' + total_split[1])
        else:
            turn_bet = float(total_split[0])
        return turn_bet

def river_action():
    river_bet = 0
    if 'Folds' in line:
        return river_bet
    elif 'Checks' in line:
        return river_bet
    elif 'Bets' in line:
        value_line = line.partition('$')[2]
        total_split = re.findall('\d+', value_line)
        if len(total_split) > 1:
            river_bet = float(total_split[0] + '.' + total_split[1])
        else:
            river_bet = float(total_split[0])
        return river_bet
    elif 'Raises' in line:
        value_line = line.partition('$')[2]
        total_split = re.findall('\d+', value_line)
        river_bet = float(total_split[0] + '.' + total_split[1])
        return river_bet
    elif 'All-in(raise)' in line:
        value_line = line.partition('$')[2]
        total_split = re.findall('\d+', value_line)
        river_bet = float(total_split[0] + '.' + total_split[1])
        return river_bet
    elif 'All-in' in line:
        value_line = line.partition('$')[2]
        total_split = re.findall('\d+', value_line)
        if len(total_split) > 1:
            river_bet = float(total_split[0] + '.' + total_split[1])
        else:
            river_bet = float(total_split[0])
        return river_bet
    elif 'Calls' in line:
        value_line = line.partition('$')[2]
        total_split = re.findall('\d+', value_line)
        if len(total_split) > 1:
            river_bet = float(total_split[0] + '.' + total_split[1])
        else:
            river_bet = float(total_split[0])
        return river_bet
    
def final_total():
    value_line = line.partition('$')[2]
    total_split = re.findall('\d+', value_line)
    if '.' in line:
        winning_total = float(total_split[0] + '.' + total_split[1])
    elif 'lost' in line or 'Mucked' in line:
        winning_total = 0
    elif '''[Does not show]''' in line and len(total_split) == 0:
        winning_total = 0
    elif '.' not in line:
        winning_total = float(total_split[0])
    else:
        winning_total = 0
    return winning_total
```

```python
seat_assigned = False
total_amount = 0
card_assign_section = False
preflop_section = False
flop_section = False
turn_section = False
river_section = False
hand_over = False
hands_played = {}
i = 0
hand_dict = {}
for file in os.listdir(HANDS_DIRECTORY):
    if 'Tourney' in file:
        pass
    else:
        hand = open(f'{HANDS_DIRECTORY}/{file}', 'r')
        hand = hand.readlines()
        for line in hand:
            if 'uncalled portion' in line and 'ME' in line:
                hand_over = True
                card_assign_section = False
                preflop_section = False
                flop_section = False
                turn_section = False
                river_section = False
                value_line = line.partition('$')[2]
                total_split = re.findall('\d+', value_line)
                returned_bet = float(total_split[0] + '.' + total_split[1])
                total_amount -= returned_bet
            elif 'Seat stand' in line:
                pass
            elif 'Table leave user' in line:
                pass
            elif 'Table enter user' in line:
                pass
            elif 'Seat sit out' in line:
                pass
            elif 'Seat sit down' in line:
                pass
            elif 'Table deposit' in line:
                pass
            elif 'Enter(Auto)' in line:
                pass
            elif 'Leave(Auto)' in line:
                pass
            elif 'Seat re-join' in line:
                pass
            else:
                if 'Ignition Hand' in line or 'Bovada Hand' in line:
                    i += 1
                    bb_value = 0
                    total_amount = 0
                    assign_section = True
                    card_assign_section = False
                    preflop_section = False
                    seat_assigned = False
                    hand_over = False
                    blinds_section = False
                if 'Small Blind  [ME] : Small Blind' in line or 'Small Blind : Small Blind' in line or 'Dealer : Set dealer' in line or 'Dealer  [ME] : Set dealer' in line or 'Big Blind : Big blind' in line or 'Big Blind  [ME] : Big blind' in line:
                    assign_section = False
                    blinds_section = True
                if assign_section:
                    my_seat = assign_seat()
                    if len(my_seat) > 0:
                        assign_section = False
                if blinds_section:
                    blinds_amount = blinds()
                    find_bb = find_BB()
                    if find_bb:
                        bb_value = find_bb
                    total_amount += blinds_amount
                    blinds_amount = 0
                if 'HOLE CARDS' in line:
                    blinds_section = False
                    card_assign_section = True
                if 'ME' in line and 'Leave' in line:
                    card_assign_section = False
                if card_assign_section:
                    if 'Folds' in line or 'Checks' in line or 'Raises' in line or 'Table' in line or 'Calls'  in line or 'All-in' in line:
                        card_assign_section = False
                        preflop_section = True
                    elif 'ME' in line:
                        hole_cards = card_assignment()
                if preflop_section:
                    if 'FLOP' in line or 'SUMMARY' in line:
                        preflop_section = False
                        if 'FLOP' in line:
                            flop_section = True
                        elif 'SUMMARY' in line:
                            hand_over = True
                    else:
                        if 'ME' in line:
                            preflop_amount = preflop_action()
                            total_amount += preflop_amount

                if flop_section:
                    if 'TURN' in line or 'SUMMARY' in line:
                        flop_section = False
                        if 'TURN' in line:
                            turn_section = True
                        elif 'SUMMARY' in line:
                            hand_over = True
                    else:
                        if 'ME' in line:
                            flop_amount = flop_action()
                            total_amount += flop_amount

                if turn_section:
                    if 'RIVER' in line or 'SUMMARY' in line:
                        turn_section = False
                        if 'RIVER' in line:
                            river_section = True
                        elif 'SUMMARY' in line:
                            hand_over = True
                    else:
                        if 'ME' in line:
                            turn_amount = turn_action()
                            total_amount += turn_amount
                if river_section:
                    if 'SUMMARY' in line:
                        river_section = False
                        hand_over = True
                    else:
                        river_action
                if hand_over and 'SUMMARY' not in line:
                    if my_seat in line:
                        if 'Folded' in line:
                            winning_total = 0
                        else:
                            winning_total = final_total()
                        profit = winning_total - total_amount
                        profit_bb = profit / bb_value
                        hole_cards_backwards = hole_cards[1::-1] + hole_cards[-1]
                        if hole_cards in hand_dict:
                            hand_dict[hole_cards] += profit_bb
                            hands_played[hole_cards ] += 1
                        elif hole_cards_backwards in hand_dict:
                            hand_dict[hole_cards_backwards] += profit_bb
                            hands_played[hole_cards_backwards] += 1
                        else:
                            hand_dict[hole_cards] = profit_bb
                            hands_played[hole_cards] = 1
```

```python
hand_df = pd.DataFrame(hand_dict.items(), columns=['Hand', 'Profit_bb'])
```

```python
hand_df = hand_df.reset_index()
hand_df = hand_df.drop(['index'], axis = 1)
```

```python
hands_played_df = pd.DataFrame(hands_played.items(), columns=['Hand', 'times_dealt'])
```

```python
hand_df['Times_dealt'] = hands_played_df['times_dealt']
```

```python
hand_df['bb_ev_per_hand'] = hand_df['Profit_bb'] / hand_df['Times_dealt']
```

```python
hand_df = hand_df.sort_values(by=['bb_ev_per_hand'])
```
